**Dosificación de Mezcla para Columnas**

Planteamiento:

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

Se sabe que para un concreto estándar de 210 kg/cm2 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos**:

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$ v=π * r^2 * h $$ V=π×r2×h
. (Sugerencia: usar 3.1416
 o importar math.pi).
4. Calcular la cantidad de sacos:
$$sacos = v * 9.5$$

**Prueba del programa**:

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3

Sacos necesarios ≈3.58

 (el estudiante verá aquí la importancia de los tipos decimales).



````python
# ENTRADA: Solicitar la altura de la columna en metros.
# Solicitar el diámetro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

# PROCESO:
r = d/2
v = 3.1416 * r** 2 *h
sacos = v * 9.5

# SALIDA:
print("Volumen ≈", v, "m3")
print("Sacos necesarios ≈", sacos)
````

A continuación, se explica cada línea del código:

```python
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))
```
Estas dos líneas solicitan al usuario que ingrese la altura y el diámetro de la columna, respectivamente. La función `input()` lee la entrada del usuario como una cadena de texto, y `float()` la convierte a un número decimal (`float`) para poder realizar cálculos.

```python
r = d/2
```
Esta línea calcula el radio (`r`) de la columna dividiendo el diámetro (`d`) ingresado por el usuario entre 2.

```python
v = 3.1416 * r** 2 *h
```
Aquí se calcula el volumen (`v`) de la columna. Se utiliza la fórmula para el volumen de un cilindro: \(V = \pi \cdot r^2 \cdot h\). Se usa `3.1416` como una aproximación de \(\pi\), `r**2` para el radio al cuadrado, y `h` para la altura.

```python
sacos = v * 9.5
```
Esta línea calcula la cantidad total de sacos de cemento necesarios (`sacos`). Se multiplica el volumen calculado (`v`) por la constante de 9.5 sacos por metro cúbico, que es la cantidad de cemento requerida por cada metro cúbico de mezcla.

```python
print("Volumen ≈", v, "m3")
print("Sacos necesarios ≈", sacos)
```
Finalmente, estas dos líneas imprimen los resultados al usuario. La primera muestra el volumen calculado de la columna en metros cúbicos, y la segunda muestra la cantidad de sacos de cemento necesarios.
