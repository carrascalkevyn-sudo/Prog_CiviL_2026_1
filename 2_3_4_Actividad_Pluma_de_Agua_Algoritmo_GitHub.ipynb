> **¿Por qué los costeños le decimos “pluma” a la llave del agua?**  
>  
> En el Caribe colombiano, abrir la “pluma” no es solo una acción cotidiana: es una expresión cargada de historia, técnica y cultura. Aunque algunos se sorprendan o incluso se burlen del término, lo cierto es que decir “pluma” en vez de “llave” no es un error, sino una herencia viva que conecta nuestro lenguaje con siglos de evolución hidráulica.

Aproximadamente, una pluma equivalía a **0.025 litros por segundo**. En algunos lugares, ese flujo se regulaba mediante orificios calibrados y dispositivos llamados **plumeros**, que permitían controlar la salida del agua.  
>  
> Durante la época colonial, muchas de estas prácticas técnicas cruzaron el Atlántico y llegaron a ciudades como Cartagena, Barranquilla y Santa Marta. Con el tiempo, el término “pluma” dejó de ser solamente técnico y pasó al habla cotidiana como sinónimo de la llave del agua.

## 2. Situación problema

Un hogar recibe una cierta cantidad de **plumas de agua**.  
Se sabe que:

- **1 pluma = 0.025 litros por segundo**

Con esta información, el estudiante debe construir un programa que permita calcular:

1. el caudal total en litros por segundo;
2. el consumo diario en litros;
3. el consumo mensual en litros (suponiendo 30 días);
4. una clasificación del consumo:
   - **Bajo**: menos de 1000 litros por día
   - **Medio**: entre 1000 y 3000 litros por día
   - **Alto**: más de 3000 litros por día

---

## 3. Análisis del problema

### 3.1 Entrada del sistema
El programa debe recibir:

- número de plumas asignadas al hogar.

### 3.2 Proceso
El programa debe:

- convertir el número de plumas en caudal total;
- transformar el caudal a consumo diario;
- transformar el consumo diario a consumo mensual;
- clasificar el nivel de consumo.

### 3.3 Salida
El programa debe mostrar:

- consumo mensual;
- clasificación del consumo.

## 5. Algoritmo propuesto

## 6. Espacio para que el estudiante construya la solución

### Instrucciones
Complete el siguiente bloque de código a partir del algoritmo anterior.

````python
# ==========================================================
# ACTIVIDAD: MODELAR LA "PLUMA DE AGUA" CON PYTHON
# ==========================================================

# 1. Solicitar al usuario el número de plumas
plumas = float(input("Ingrese el número de plumas: "))

# 2. Definir constantes del problema
caudal_pluma = 0.025  # litros por segundo
segundos_dia = 24 * 60 * 60  # segundos en un día

# 3. Calcular el caudal total
caudal_total = plumas * caudal_pluma

# 4. Calcular el consumo diario
consumo_diario = caudal_total * segundos_dia

# 5. Calcular el consumo mensual
consumo_mensual = consumo_diario * 30  # Suponiendo 30 días en el mes

# 6. Clasificar el consumo
if consumo_diario < 1000:
    nivel_consumo = "Bajo"
elif 1000 <= consumo_diario <= 3000:
    nivel_consumo = "Medio"
else:
    nivel_consumo = "Alto"

# 7. Mostrar resultados
print(f"\n--- Resultados ---")
print(f"Caudal total: {caudal_total:.4f} litros/segundo")
print(f"Consumo diario: {consumo_diario:.2f} litros")
print(f"Consumo mensual: {consumo_mensual:.2f} litros (estimado para 30 días)")
print(f"Clasificación del consumo: {nivel_consumo}")
````

# Define el caudal que representa una sola 'pluma de agua' en litros por segundo.
caudal_pluma = 0.025  # litros por segundo

consumo_diario = caudal_total * segundos_dia

*   Las estructuras de control utilizadas y por qué eran apropiadas.
*   Puntos de mejora en cuanto a robustez (manejo de errores de entrada) y modularidad (uso de funciones).
*   Finalmente, la versión comentada del código es una guía excelente para futuras revisiones o para que otros entiendan rápidamente la lógica.

### ¿Qué importancia tiene publicar y versionar su trabajo?
Publicar y versionar el trabajo es fundamental por varias razones:
*   **Colaboración:** Permite que otros vean y contribuyan al proyecto, fomentando el aprendizaje mutuo.
*   **Historial de Cambios:** El control de versiones (Git/GitHub) crea un registro detallado de todas las modificaciones, lo que es esencial para rastrear errores, revertir cambios o entender la evolución del código.
*   **Portafolio:** Un repositorio bien documentado en GitHub sirve como un portafolio profesional, mostrando habilidades de programación y capacidad de gestión de proyectos.
*   **Acceso y Respaldo:** Proporciona un lugar seguro para almacenar el código, accesible desde cualquier lugar y actuando como un respaldo ante cualquier eventualidad local.
