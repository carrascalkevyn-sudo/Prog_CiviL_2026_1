## Teoría del Confort Térmico en Ganado Vacuno

El confort térmico en el ganado vacuno es un estado de equilibrio en el que los animales no experimentan estrés por calor ni por frío. Mantener este equilibrio es crucial para la salud, bienestar y productividad del ganado, especialmente en la producción lechera y cárnica. Cuando las vacas se desvían de su zona de termoneutralidad, activan mecanismos fisiológicos para disipar o conservar calor, lo que conlleva un gasto energético que podría destinarse a la producción.

### Índice de Temperatura y Humedad (ITH)

Uno de los indicadores más utilizados para evaluar el estrés por calor en el ganado es el **Índice de Temperatura y Humedad (ITH)**. Este índice combina la temperatura del aire y la humedad relativa para proporcionar una medida única de la carga térmica ambiental que experimentan los animales. La fórmula utilizada comúnmente es:

$ITH = 0.72 * (T_{bs} + T_{bh}) + 40.6$

Donde:
1. $T_{bs}$: Temperatura de bulbo seco (°C)
2. $T_{bh}$: Temperatura de bulbo húmedo (°C)

### Interpretación del ITH para Ganado Bovino (según National Weather Service):

La interpretación de los valores de ITH permite clasificar el nivel de estrés térmico que puede afectar al ganado y, por ende, su producción de leche. Los umbrales comúnmente aceptados son:

*   **ITH ≤ 70**: Ambiente no estresante. El ganado se encuentra en su zona de confort térmico y no se espera una afectación significativa en su bienestar o productividad.
*   **ITH > 70 y ≤ 78**: Situación de alerta por estrés térmico. Los animales pueden empezar a mostrar signos de estrés leve a moderado. Se recomienda implementar medidas preventivas.
*   **ITH > 78 y ≤ 83**: Peligro por estrés térmico. El ganado está bajo un estrés considerable, lo que puede llevar a una disminución notable en la producción y aumentar el riesgo de problemas de salud. Se requieren intervenciones activas.
*   **ITH > 83**: Situación crítica. Los animales experimentan un estrés térmico severo. Existe un alto riesgo de golpes de calor, enfermedades graves y mortalidad. Se necesitan medidas de mitigación urgentes y agresivas.

### Impacto del Estrés Térmico

El estrés térmico en el ganado vacuno puede generar una serie de impactos negativos, incluyendo:

*   **Disminución de la producción**: Reducción en la producción de leche, menor ganancia de peso y menor eficiencia alimenticia.
*   **Problemas reproductivos**: Reducción de la fertilidad, aumento de abortos y ciclos estrales irregulares.
*   **Salud animal**: Debilitamiento del sistema inmune, aumento de la susceptibilidad a enfermedades y acidosis.
*   **Comportamiento**: Alteración en los patrones de alimentación, descanso y rumia.

# **Metodos y calculos**
• Índice de temperatura y humedad (producción intensiva)


$𝑇𝐻 = 0.72 * (𝑇{𝑏𝑠} +𝑇{𝑏ℎ})+40.6 $

donde:
1. $ T _ {bs}$ Temperatura bulbo seco (°C)
2. $T_{bh}$ Temperatura de bulbo húmedo (°C)

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda
utilizar los límites propuestos por National Weather Service:


1. Sí $ITH ≤ 70 $ ambiente no estresante
2. Sí $ITH > 70 \space and \space ≤ 78$  situación de alerta por estrés térmico
3. Sí $ITH > 78 \space and \space ≤ 83$ peligro por estrés térmico
4. Sí $ITH > 83$  situación crítica

````python
#ENTRADA: pedir tbs (Temperatura del bulbo seco) y tbh (Temoeratura del bulbo humedo)
tbs = float(input("Ingrese la temperatura bulbo seco: "))
tbh = float(input("Ingrese la temperatura bulbo húmedo: "))

#PROCESO: Calculo de ith (indice de temperatura y humedad)
ith= 0.72 * (tbs + tbh) + 40.6

#SALIDA: imprimir resultado de la evaluacion de ith y el confort termico
print("El indice de temperatura y humedad es: ", ith)
print("El confort térmico es: ")

if ith <= 70:
  print("Ambiente no estresante")
elif ith > 70 and ith <= 78:
  print("Situación de alerta por estrés térmico")

else:
  print("Situación crítica")
````
