# Problema de Ingeniería (Nivel Básico)

### Cálculo del **área de un terreno rectangular**

Este problema es muy común en **ingeniería civil, agrícola y agroindustrial**, por ejemplo para:

* calcular el área de un lote agrícola
* estimar superficie de cultivo
* planificar riego o construcción

---

# 1. Definición del algoritmo

## Entradas

Datos que el usuario debe ingresar.

* **Largo del terreno (m)**
* **Ancho del terreno (m)**

## Proceso

Operación matemática necesaria.

[
Area = Largo \times Ancho
]

## Salida

Resultado que el programa mostrará.

* **Área del terreno en metros cuadrados**

---

# 2. Algoritmo (explicación paso a paso)

1. Solicitar al usuario el **largo del terreno**.
2. Solicitar el **ancho del terreno**.
3. Multiplicar **largo × ancho**.
4. Guardar el resultado en la variable **area**.
5. Mostrar el **área calculada**.

---

# 3. Pseudocódigo

```
INICIO

Leer largo
Leer ancho

area = largo * ancho

Mostrar area

FIN
```

---

# 4. Implementación en Python

```python
# -----------------------------------------
# Programa: Cálculo del área de un terreno
# Autor: Estudiante de Ingeniería
# -----------------------------------------

# Entrada de datos
largo = float(input("Ingrese el largo del terreno (m): "))
ancho = float(input("Ingrese el ancho del terreno (m): "))

# Proceso
area = largo * ancho

# Salida
print("El área del terreno es:", area, "metros cuadrados")
```

---

# Ejemplo de ejecución

```
Ingrese el largo del terreno (m): 50
Ingrese el ancho del terreno (m): 20

El área del terreno es: 1000 metros cuadrados
```
