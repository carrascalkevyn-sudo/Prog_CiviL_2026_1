# DISEÑO DE CANAL TRAPEZOIDAL DE RIEGO
# Programación Aplicada - Funciones y Estructuras

Contexto
Un equipo de ingenieros civiles y agroindustriales debe diseñar un canal trapezoidal de riego para abastecer cultivos en una zona rural. El canal debe suministrar suficiente caudal para distintas demandas de agua de los cultivos, y además operar en un régimen hidráulico seguro (subcrítico), que evita erosión y daños en la estructura.

# Marco teórico
Ecuación de Manning — relaciona la geometría del canal con el caudal transportado:

$$ Q=\frac{1}{n} * A * R^\frac{2}{3} * S^\frac{1}{2} $$

Donde: - Q = caudal (m³/s) - n = coeficiente de rugosidad de Manning (adimensional) - A = área de la sección transversal (m²) - R = radio hidráulico (m) = A / P - S = pendiente longitudinal del canal (m/m)

Geometría del canal trapezoidal:

$$A=(b+z⋅y)⋅y$$  $$P=b+2y\sqrt{1 + z^2} $$    $$T=b+2⋅z⋅y$$

Donde b = base, y = tirante (profundidad del agua), z = talud lateral, T = ancho superficial.

Número de Froude — clasifica el régimen de flujo:

$$Fr=\frac{V}{\sqrt{g * Dh}}   con  Dh=\frac{A}{T},V=\frac{Q}{A} $$

#Condición	Régimen

Fr < 1	Subcrítico (flujo tranquilo) ✅ recomendado para riego

Fr = 1	Crítico

Fr > 1	Supercrítico (flujo rápido) ⚠️ riesgo de erosión

#Demanda hídrica agroindustrial — Cada cultivo tiene una demanda en mm/día, que se convierte a caudal (m³/s) según el área sembrada:

$$Qcultivo=\frac{ETo *  Kc * A_cultivo}{86400}$$

Donde ETo = evapotranspiración de referencia (mm/día), Kc = coeficiente de cultivo, A_cultivo = área en m².

```python
# DISEÑO DE CANAL TRAPEZOIDAL DE RIEGO
# Programación Aplicada - Funciones y Estructuras

import math

def calcular_area(b, z, y):
    """
    Calcula el área hidráulica de un canal trapezoidal.

    Parámetros:
    b : base del canal (m)
    z : talud lateral
    y : tirante hidráulico (m)

    Retorna:
    Área hidráulica (m²)
    """
    return (b + z * y) * y

def calcular_perimetro(b, z, y):
    """
    Calcula el perímetro mojado de un canal trapezoidal.

    Retorna:
    Perímetro mojado (m)
    """
    return b + 2 * y * math.sqrt(1 + z**2)


def calcular_ancho_superficial(b, z, y):
    """
    Calcula el ancho superficial de un canal trapezoidal.

    Retorna:
    Ancho superficial (m)
    """
    return b + 2 * z * y
```

```python
def calcular_caudal(b, z, y, n, S):
    """
    Calcula el caudal y la velocidad usando la ecuación de Manning.

    Parámetros:
    b : base del canal (m)
    z : talud lateral
    y : tirante hidráulico (m)
    n : coeficiente de Manning
    S : pendiente longitudinal

    Retorna:
    Q : caudal (m³/s)
    V : velocidad (m/s)
    """

    # Verificación de datos positivos
    if b <= 0 or z <= 0 or y <= 0 or n <= 0 or S <= 0:
        print("Error: todos los parámetros deben ser positivos.")
        return None, None

    A = calcular_area(b, z, y)
    P = calcular_perimetro(b, z, y)

    R = A / P

    Q = (1 / n) * A * (R ** (2/3)) * (S ** 0.5)

    V = Q / A

    return Q, V

def clasificar_regimen(Fr):
    """
    Clasifica el régimen de flujo según el número de Froude.

    Parámetro:
    Fr : número de Froude

    Retorna:
    String con el tipo de régimen
    """

    if Fr < 1:
        return "Subcrítico "
    elif Fr == 1:
        return "Crítico"
    else:
        return "Supercrítico Riesgo de erosión"

```

```python
import math

def analizar_canal(b, z, n, S, y_min, y_max, paso):
    """
    Analiza el comportamiento hidráulico del canal
    para distintos valores de tirante.

    Retorna:
    Lista de diccionarios con resultados.
    """

    resultados = []

    y = y_min

    while y <= y_max:

        Q, V = calcular_caudal(b, z, y, n, S)

        A = calcular_area(b, z, y)
        T = calcular_ancho_superficial(b, z, y)

        Dh = A / T

        Fr = V /math.sqrt(9.81 * Dh)

        regimen = clasificar_regimen(Fr)

        resultados.append({
            "tirante": round(y, 2),
            "area": round(A, 3),
            "caudal": round(Q, 3),
            "velocidad": round(V, 3),
            "froude": round(Fr, 3),
            "regimen": regimen
        })

        y += paso

    return resultados
```

```python
def calcular_demanda_cultivo(nombre, ETo, Kc, area_ha):
    """
    Calcula la demanda hídrica de un cultivo.

    Parámetros:
    nombre : nombre del cultivo
    ETo : evapotranspiración (mm/día)
    Kc : coeficiente de cultivo
    area_ha : área sembrada (hectáreas)

    Retorna:
    Diccionario con la demanda del cultivo
    """

    area_m2 = area_ha * 10000

    Q_cultivo = (ETo * Kc * area_m2) / 86400 / 1000

    return {
        "nombre": nombre,
        "demanda": Q_cultivo
    }


def verificar_abastecimiento(Q_canal, cultivos):
    """
    Verifica si el caudal del canal abastece
    la demanda de cada cultivo.
    """

    for cultivo in cultivos:

        print("\nCultivo:", cultivo["nombre"])
        print("Demanda:", round(cultivo["demanda"], 4), "m³/s")

        if Q_canal >= cultivo["demanda"]:
            print(" El canal abastece el cultivo.")
        else:
            print(" El canal NO abastece el cultivo.")
```

```python
# PROGRAMA PRINCIPAL
# Parámetros del canal
b = 0.80
z = 1.5
n = 0.014
S = 0.0008

# Análisis hidráulico
resultados = analizar_canal(b, z, n, S, 0.20, 1.20, 0.20)

print(f"\n{'y(m)':>6} {'A(m²)':>7} {'Q(m³/s)':>9} {'Fr':>7} {'Régimen':>20}")
print("-" * 60)

for r in resultados:
    print(
        f"{r['tirante']:>6.2f} "
        f"{r['area']:>7.3f} "
        f"{r['caudal']:>9.4f} "
        f"{r['froude']:>7.3f} "
        f"{r['regimen']:>20}"
    )

# Cultivos
cultivos_data = [
    {"nombre": "Maíz",  "ETo": 5.2, "Kc": 1.05, "area_ha": 40},
    {"nombre": "Arroz", "ETo": 6.0, "Kc": 1.20, "area_ha": 350},
    {"nombre": "Palma", "ETo": 5.5, "Kc": 1.10, "area_ha": 600},
    {"nombre": "Caña",  "ETo": 5.8, "Kc": 1.25, "area_ha": 800},
]

cultivos = []
for cultivo in cultivos_data:
    demanda_info = calcular_demanda_cultivo(cultivo['nombre'], cultivo['ETo'], cultivo['Kc'], cultivo['area_ha'])
    cultivos.append(demanda_info)

# Caudal de diseño
Q_diseño = 0.663   # m³/s para y = 0.80 m

print("\n")
print("=" * 60)
print("VERIFICACIÓN DE ABASTECIMIENTO")
print("=" * 60)

verificar_abastecimiento(Q_diseño, cultivos)
```
